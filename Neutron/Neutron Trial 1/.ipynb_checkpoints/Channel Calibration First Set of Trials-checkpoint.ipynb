```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

#data = np.loadtxt(tsv_file, delimiter="\t", skiprows=1, usecols=[0,1,2])

data_start = 22
calibration_na22_t1 = "Na22Cal16xGain.tsv"
calibration_na22_t2 = "Na22Cal16xGain_LLDSet.tsv"
calibration_co60_t1 = "Co60Cal16xGain_LLDSet.tsv"

_2_block_paraffin = "2xLead_HalfParaffin.tsv"

na22_t1 = np.loadtxt( calibration_na22_t1 , unpack=True, skiprows = data_start, delimiter = "\t\t") 
na22_t2 = np.loadtxt( calibration_na22_t2 , unpack=True, skiprows = data_start, delimiter = "\t\t") 
co60_t1 = np.loadtxt( calibration_co60_t1 , unpack=True, skiprows = data_start, delimiter = "\t\t") 

#data = np.loadtxt(calibration_na22_t2, delimiter="\t\t", skiprows=data_start, usecols=[0,1,2])

#tsv_file = open("Na22Cal16xGain_LLDSet _copy.tsv")
#read_tsv = csv.reader(tsv_file, delimiter="\t\t")

channel_na22_t1 = na22_t1[0]
N_na22_t1 = na22_t1[1]
dN_na22_t1 = np.sqrt(N_na22_t1)
for i, value in enumerate(dN_na22_t1):
    if value == 0:
        dN_na22_t1[i] = 1.14
        

channel_na22_t2 = na22_t2[0]
N_na22_t2 = na22_t1[1]
dN_na22_t2 = np.sqrt(N_na22_t2)
for i, value in enumerate(dN_na22_t2):
    if value == 0:
        dN_na22_t2[i] = 1.14
        
channel_co60_t1 = co60_t1[0]
N_co60_t1 = co60_t1[1]
dN_co60_t1 = np.sqrt(N_co60_t1)
for i, value in enumerate(dN_co60_t1):
    if value == 0:
        dN_co60_t1[i] = 1.14
        
fig,ax = plt.subplots(figsize = (12,8))
ax.errorbar(channel_na22_t1, N_na22_t1, dN_na22_t1, fmt='k.')

fig,ax = plt.subplots(figsize = (12,8))
ax.errorbar(channel_na22_t2, N_na22_t2, dN_na22_t2, fmt='k.')

fig,ax = plt.subplots(figsize = (12,8))
ax.errorbar(channel_co60_t1, N_co60_t1, dN_co60_t1, fmt='k.')
```

```python
# so now we want to find the best values for the channels for 511kev
# and 1.27 MeV for Na-22 source
# and 1.1732 MeV and 1.3325 MeV for co-60
# to do so we start with the easiest convergence fit for 511 kev

min_value = 110
max_value = 147

fig,ax = plt.subplots(figsize = (12,8))
ax.errorbar(channel_na22_t1[min_value:max_value], N_na22_t1[min_value:max_value], dN_na22_t1[min_value:max_value], fmt='k.')
```

```python
def gaussianfunc_lin(p,x):
    return p[0]/(p[2]*np.sqrt(2*np.pi))*np.exp(-(x-p[1])**2/(2*p[2]**2)) + p[3]*x +p[4]

def residual(p,func, xvar, yvar, err):
    return (func(p, xvar) - yvar)/err

#channel_T1_0, N_T1_0, dN_T1_0

# The code below defines our data fitting function.
# Inputs are:
# initial guess for parameters p0
# the function we're fitting to
# the x,y, and dy variables
# tmi can be set to 1 or 2 if more intermediate data is needed

def data_fit(p0,func,xvar, yvar, err,tmi=0):
    try:
        fit = optimize.least_squares(residual, p0, args=(func,xvar, yvar, err),verbose=tmi)
    except Exception as error:
        print("Something has gone wrong:",error)
        return p0,np.zeros_like(p0),-1,-1
    pf = fit['x']

    print()

    try:
        cov = np.linalg.inv(fit['jac'].T.dot(fit['jac']))          
        # This computes a covariance matrix by finding the inverse of the Jacobian times its transpose
        # We need this to find the uncertainty in our fit parameters
    except:
        # If the fit failed, print the reason
        print('Fit did not converge')
        print('Result is likely a local minimum')
        print('Try changing initial values')
        print('Status code:', fit['status'])
        print(fit['message'])
        return pf,np.zeros_like(pf),-1,-1
            #You'll be able to plot with this, but it will not be a good fit.

    chisq = sum(residual(pf,func,xvar, yvar, err) **2)
    dof = len(xvar) - len(pf)
    red_chisq = chisq/dof
    pferr = np.sqrt(np.diagonal(cov)) # finds the uncertainty in fit parameters by squaring diagonal elements of the covariance matrix
    print('Converged with chi-squared {:.2f}'.format(chisq))
    print('Number of degrees of freedom, dof = {:.2f}'.format(dof))
    print('Reduced chi-squared {:.2f}'.format(red_chisq))
    print()
    Columns = ["Parameter #","Initial guess values:", "Best fit values:", "Uncertainties in the best fit values:"]
    print('{:<11}'.format(Columns[0]),'|','{:<24}'.format(Columns[1]),"|",'{:<24}'.format(Columns[2]),"|",'{:<24}'.format(Columns[3]))
    for num in range(len(pf)):
        print('{:<11}'.format(num),'|','{:<24.3e}'.format(p0[num]),'|','{:<24.3e}'.format(pf[num]),'|','{:<24.3e}'.format(pferr[num]))
    return pf, pferr, chisq,dof
```

```python
channel_na22_t1_2 = channel_na22_t1[min_value:max_value]
N_na22_t1_2 = N_na22_t1[min_value:max_value]
dN_na22_t1_2 = dN_na22_t1[min_value:max_value]

p0 = [9.976e+05, 1.301e+02, 7.710e+00 , -8.874e+01 , 1.597e+04]

pf1, pferr1, chisq1, dof1 = data_fit(p0,gaussianfunc_lin, channel_na22_t1_2, N_na22_t1_2, dN_na22_t1_2)
```
